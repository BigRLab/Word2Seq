# Curve Fit the Loss Function

## Function

$$
f(x) = Ae^{-Bx} + C
$$

```python
def f(x, a, b, c):
    return a * np.exp(-b * x) + c
```

## Fit the Parameters

```python
x = np.arange(len(df.val_loss.values))
y = df.val_loss.values

(a, b, c), err = sp.optimize.curve_fit(f,  x,  y)
```

## Check Error

```python
print(a, b, c)
print(err)
```
